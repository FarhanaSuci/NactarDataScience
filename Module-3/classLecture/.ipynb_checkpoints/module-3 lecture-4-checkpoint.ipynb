```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y,'^-b',y,x,'o-g')
plt.xlabel("x-axis")
plt.ylabel("y axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x_data=[1,2,3,4,5,6]
y_data=[9,10,11,25,26,27]
plt.plot(x_data,y_data, 'g',y_data, x_data, 'o-g', linewidth=5)
plt.xlabel("x_data")
plt.ylabel("y_data")
plt.show()
```

```python
x=np.linspace(0,10,100)
#print(x)
y1=x**2
y2=x**3
y3=x*2.5
plt.plot(x,y1,label="square")
plt.plot(x,y2,label="cube")
plt.plot(x,y3,label="mulby_2.5")
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('tt')
plt.grid(True)
plt.legend(loc=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
y1 = np.random.randn(70)
y2 = np.random.randn(70)
# Both y1 and y2 should be of same size
plt.scatter(y1, y2)
plt.axis([-3,3, -3, 3])
plt.show()
```

```python
x = np.arange(70)
y1 = np.random.randn(70)
y2 = np.random.randn(70)
plt.scatter(x, y1, marker ='o', label="y1 with x",color='r')
plt.scatter(x, y2, marker ='v', label="y2 with x",color='g')
plt.legend(loc = 'upper left')
plt.show()
```

```python
age_group = ['<18','19-30','31-50','51-70', '>70']
total_num = [12, 18, 17, 8, 6]  # value in multiple of 100000
plt.bar(age_group, total_num)
plt.xlabel("Age Group")
plt.ylabel("Population x100000")
plt.title("Population Stats for Bangalore")
plt.show()
```

```python
age_group = ['<18','19-30','31-50','51-70', '>70']
total_num = [12, 18, 17, 8, 6]  # value in multiple of 100000
plt.bar(age_group, total_num, color = 'g', edgecolor= 'r', width =.4) #default is 0.8
plt.xlabel("Age Group")
plt.ylabel("Population x100000")
plt.title("Population Stats for Bangalore")
plt.show()
```

```python
age_group = ['<16','19-30','31-50','51-70', '>70']
total_num = [12, 18, 17, 8, 6]  # value in multiple of 100000
colors = ['#84f011', '#ed3d0c', 'b', 'c', 'y']
plt.bar(age_group, total_num, color = colors, width =0.4)
plt.xlabel("Age Group")
plt.ylabel("Population x100000")
plt.title("Population Stats for Bangalore")
plt.show()
```
