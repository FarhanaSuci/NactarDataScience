```python

t1=np.arange(0.0, 5.0, 0.1)
y1=np.sin(2*np.pi*t1)
y2=np.cos(2*np.pi*t1)
y3=np.tan(2*np.pi*t1)

plt.figure()
plt.subplot(311)
plt.plot(t1,y1,'r')
plt.subplot(312)
plt.plot(t1,y2,'g')
plt.subplot(313)
plt.plot(t1,y3,'g')
plt.show()
```

```python
t1=np.arange(0.0, 5.0, 0.1)
y1=np.sin(2*np.pi*t1)
y2=np.cos(2*np.pi*t1)
y3=np.tan(2*np.pi*t1)

plt.figure(1)
plt.subplot(131)
plt.plot(t1,y1,'r')
plt.subplot(132)
plt.plot(t1,y2,'g')
plt.grid(True)
plt.subplot(133)
plt.plot(t1,y3,'g')
plt.grid(True)

plt.figure(2)
plt.subplot(311)
plt.plot(t1,y1,'r')
plt.subplot(312)
plt.plot(t1,y2,'g')
plt.subplot(313)
plt.plot(t1,y3,'g')
plt.grid(True)

plt.show()
```

stack chart

```python

avg_score = [50, 20, 57, 60, 26, 55, 88, 19, 74, 60]
sub = ['Math','Phy','Chem','Bio','IT', 'Sci', 'Tec','PT','ECE', 'Comp']
grace_marks = [12, 24, 6, 5, 7, 10, -7, 20, 4, 9]

plt.title('Average score for Subjects', fontsize =16)
plt.xlabel('Subjects',fontsize = 14, color = 'm')
plt.ylabel('Scores', fontsize = 14, color = 'g')

plt.bar(sub,avg_score, label ='Average Score' )
plt.bar(sub, grace_marks , bottom = avg_score)

plt.show()
```
